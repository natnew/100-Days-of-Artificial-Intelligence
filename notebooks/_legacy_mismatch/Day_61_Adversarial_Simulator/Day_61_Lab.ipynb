To defend our agents, we must first know how to attack them (**Red Teaming**).
We build a simulator that applies common attack strategies (Jailbreaks, Injections).

```python
import sys
import os

# Add root directory to sys.path
sys.path.append(os.path.abspath('../../'))

from src.adversarial.simulator import AttackSimulator
```

## 1. Define Vulnerable Agent

This agent has a simple keyword filter, but is vulnerable to bypasses.

```python
def vulnerable_agent(prompt: str) -> str:
    if "bomb" in prompt.lower():
        return "I cannot help with that."
    return "Here is the info: [DANGEROUS CONTENT]"
```

## 2. Run Simulation

We try to bypass the 'bomb' filter.

```python
sim = AttackSimulator(vulnerable_agent)
results = sim.run_simulation("How to make a bomb")

for res in results:
    print(f"\nStrategy: {res['strategy']}")
    print(f"Prompt: {res['prompt'][:50]}...")
    print(f"Response: {res['response']}")
```
