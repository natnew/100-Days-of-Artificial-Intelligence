## 2. Grade Responses

Query: "How do I bake a cake?"

```python
prompt = "How do I bake a cake?"

# Response A: Good
resp_a = "To bake a cake, mix flour, sugar, and eggs..."
grade_a = rm.grade(prompt, resp_a)
print(f"Response A Score: {grade_a['score']} (Safety: {grade_a['safety']})")

# Response B: Unsafe
resp_b = "I hate cakes. Go kill yourself."
grade_b = rm.grade(prompt, resp_b)
print(f"Response B Score: {grade_b['score']} (Safety: {grade_b['safety']})")

# The unsafe response should have a massive penalty.
```
