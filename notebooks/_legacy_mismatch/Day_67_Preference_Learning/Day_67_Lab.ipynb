# Day 67: Preference Learning (RLHF Step 2)

Once we have a Reward Model (Day 66), we need to tune it using human preferences.
We collect pairs of (Chosen, Rejected) responses and update the model to prefer the Chosen one.

```python
import sys
import os

# Add root directory to sys.path
sys.path.append(os.path.abspath('../../'))

from src.alignment.preferences import PreferenceDataset, PreferenceTrainer
```

## 1. Create Dataset


```python
dataset = PreferenceDataset()
# Scenario: User asks about safety.
# Chosen: "Here is a safe answer."
# Rejected: "Here is a bad answer."
dataset.add_feedback("Safety question", "This is a safe answer", "This is a bad answer")
```

## 2. Train Model

Initially, model might treat them equally if untrained.

```python
trainer = PreferenceTrainer()
print(f"Initial Weights: Safety={trainer.safety_weight}")

s_chosen = trainer.predict_score("This is a safe answer")
s_rejected = trainer.predict_score("This is a bad answer")
print(f"Scores Before: Chosen={s_chosen}, Rejected={s_rejected}")

# Train
trainer.train_step(dataset.data)

print(f"Updated Weights: Safety={trainer.safety_weight}")

s_chosen_after = trainer.predict_score("This is a safe answer")
print(f"Scores After: Chosen={s_chosen_after}")
# Expect higher score due to 'safe' weight boost.
```
