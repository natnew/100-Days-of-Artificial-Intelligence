# Day 60: Phase 3 Capstone - The White Box Agent

We combine all interpretability tools into a single agent that explains itself.
Every response includes:
1.  **Trace**: The execution path.
2.  **Attribution**: Which words mattered.
3.  **Concepts**: Abstract themes detected (e.g., Aggression).

```python
import sys
import os

# Add root directory to sys.path
sys.path.append(os.path.abspath('../../'))

from src.interpretability.white_box import WhiteBoxAgent
```

## 1. Run Agent


```python
agent = WhiteBoxAgent()
response = agent.run("I hate this process, please kill it.")

print(f"Result: {response['result']}")
```

## 2. Inspect Explanation


```python
expl = response['explanation']

print("\n--- 1. Trace ---")
for t in expl.trace:
    print(f"[{t['type']}] {t['content']}")

print("\n--- 2. Key Concepts ---")
print(expl.concepts)

print("\n--- 3. Drivers (Attribution) ---")
for token, score in expl.attributions:
    if score > 0.4:
        print(f"{token}: {score}")
```
