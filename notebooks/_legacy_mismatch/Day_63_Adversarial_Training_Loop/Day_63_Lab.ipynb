# Day 63: Adversarial Training Loop

Static defenses are not enough. We need a system that evolves.
We implement a loop: **Attack -> Analyze -> Patch**.

```python
import sys
import os

# Add root directory to sys.path
sys.path.append(os.path.abspath('../../'))

from src.adversarial.training import RobustAgent, DefenseLoop
from src.adversarial.simulator import AttackSimulator
```

## 1. Setup Environment


```python
agent = RobustAgent()
# Wrap agent.process to match simulator signature
sim = AttackSimulator(agent.process)
loop = DefenseLoop(agent, sim)
```

## 2. Run Loop (Epoch 1)

Attack with 'sensitive data'.
Agent initially allows it (vulnerability).

```python
loop.run_epoch("sensitive data")
```

## 3. Verify Patch (Epoch 2)

Try the same attack again. It should be blocked now.

```python
print("\n--- Re-running attacks ---")
res = agent.process("sensitive data")
print(f"Result: {res}")
# Should be BLOCKED if the loop worked.
```
